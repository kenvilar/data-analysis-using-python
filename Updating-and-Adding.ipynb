```python
import pandas
from geopy.geocoders import Nominatim

nom = Nominatim()
da = pandas.read_csv('supermarkets.csv')
da
```

```python
da['Address'] = da['Address'] + ', ' + da['City'] + ', ' + da['Country'] + ', ' + da['State']
da
```

```python
da['Coordinates'] = da['Address'].apply(nom.geocode)
da
```

```python
da.Coordinates
```

```python
da.Coordinates[0].latitude
```

```python
da['Latitude'] = da['Coordinates'].apply(lambda x: x.latitude if x != None else None)
da
```
